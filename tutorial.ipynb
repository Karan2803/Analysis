```python
import pandas as pd 
import numpy as np
from sklearn.datasets import make_classification
```

```python
models = {}

from sklearn.linear_model import LogisticRegression
models['Logistic Regression'] = LogisticRegression()

from sklearn.tree import DecisionTreeClassifier
models['Decision Tree Classifier'] = DecisionTreeClassifier()

from sklearn.svm import SVC
models['Support Vector Machine'] = SVC()

from sklearn.ensemble import GradientBoostingClassifier
models['Gradient Boosting Classifier'] = GradientBoostingClassifier()

from sklearn.ensemble import RandomForestClassifier
models['Random Forest Classifier'] = RandomForestClassifier()
```

```python
models.keys()
```

```python
x, y = make_classification()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy, precision, recall = {}, {}, {}

for key in models.keys():
    
    clf = models[key].fit(x_train, y_train)
    
    predictions = clf.predict(x_test)
    
    accuracy[key] = accuracy_score(y_true=y_test, y_pred=predictions)
    precision[key] = precision_score(y_true=y_test, y_pred=predictions)
    recall[key] = recall_score(y_true=y_test, y_pred=predictions)
```

```python
accuracy
```

```python
df = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])
df['Accuracy'] = accuracy.values()
df['Precision'] = precision.values()
df['Recall'] = precision.values()
```
